```python
import pandas as pd
import numpy as np
import requests
```

```python
url = 'https://www.dropbox.com/scl/fi/5x81bm78jvsotg24lax4x/CoCdata_V2_0_Posted11012024.dta?rlkey=43dg795m9m7l58vpmoggix72j&dl=1'
local_filename = url.split('/')[-1]
with requests.get(url,stream=True) as r:
    r.raise_for_status()
    with open(local_filename,'wb') as f:
        for chunk in r.iter_content(chunk_size=8192):
            f.write(chunk)
df = pd.read_stata()
df
```
