## Q1
# a

```python
df['loggdpwok'] = np.log(df['rgdpwok'])
plt.scatter(df['loggdpwok'],df['ki'])
plt.xlabel('log of GDP per worker')
plt.ylabel('real investment rate (%)')
plt.title('Figure 1: Investment rate to log GDP per worker of all countries, 2003')

coef = np.polyfit(df['loggdpwok'],df['ki'],1)
linfit = np.poly1d(coef)
plt.plot(df['loggdpwok'],linfit(df['loggdpwok']),color='red')

plt.savefig('Figure_1')
plt.show()
```

As we can see from the Figure 1 and the fit line, there is a weak positive relation between the per capita GDP of a country and its investment rates. A poor country tends to have a lower investment rate.

# b

```python
g = 0.02
delta = 0.1
df['prod_k'] = (g + delta) / df['ki']
plt.scatter(df['loggdpwok'],df['prod_k'])
plt.xlabel('log of GDP per worker')
plt.ylabel('productivity of capital')
plt.title('Figure 2: Capital productivity to log GDP per worker of all countries, 2003')

coef = np.polyfit(df['loggdpwok'],df['prod_k'],1)
linfit = np.poly1d(coef)
plt.plot(df['loggdpwok'],linfit(df['loggdpwok']),color='red')

plt.savefig('Figure_2')
plt.show()
```

Figure 2 shows a negative correlation between log per capital GDP and productivity. A poor country has higher productivity of capital. This can be explained by the fact that poorer countries, defined as lower GDP per capita, have an lower level of capital given economic growth and capital depreciation rate, and thus it has a higher level of marginal return to capital.
